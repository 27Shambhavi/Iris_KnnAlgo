```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
df.head()
```

```python
df['target']=iris.target
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.shape
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
plt.xlabel=('sepal length')
plt.ylabel=('sepal width')
plt.title('scatter plot')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='blue',label='setosa',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='green',label='versicolor',marker='*')
```

```python
plt.xlabel=('petal length')
plt.ylabel=('petal width')
plt.title('scatter plot')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='darkblue',label='setosa',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='red',label='versicolor',marker='*')
```

```python
x=df.drop(['target'],axis='columns')
x
```

```python
y=df['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
len(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
import seaborn as sj
sj.heatmap(cm,annot=True)
```
